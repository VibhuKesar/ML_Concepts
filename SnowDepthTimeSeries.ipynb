## Rolling statistics

Another approach is to take a rolling mean. Under this strategy, one chooses a "window", say $w$. Then for each point in your time series, the raw value is replaced by the average of the $w$ most recent values. This is a natural method for reducing the noise of a signal. Compared with the seasonality decomposition, it is simpler and makes fewer assumptions. However this doesn't allow one to distinguish between noise and seasonality. Also there is not an ideal way to choose the window size $w$. For these examples, $w=12$ so that the rolling mean captures the last year.

```python
pd.rolling_mean(snow_means, window=12).plot()
```

```python
pd.rolling_mean(snow_means, window=36).plot()
```

```python
## plot both to more easily compare
moving_avg = snow_means.rolling(window=12,center=False).mean()
plt.plot(snow_means)
plt.plot(moving_avg, color="red")
```
